#LISTAS

#Ciclo for


```python
estudiantes=[]
#definimos el tamaño de la lista
num_estudiantes=int(input("ingrese el numero de estudiantes: "))


#ciclo for, permite que uno ingrese datos hasta donde indique o desee

for j in range(num_estudiantes):
  print("codigo", j)
  nombre=input("ingrese el nombre: ")
  #agg los nombres por ordeen de llegada
  estudiantes.append(nombre)

  print(estudiantes)
```

CICLO WHITE

```python
#Se genera una lista vacía, que se puede ir llenando o modificando

empleados=[]
#agg datos a una lista sin tamaño definido

#bandera
e=False
#Mientras una condición sea verdad el se va a seguir repitiendo, si es falsa ya no se ejecuta
#ciclo while, se ejecuta mientras se cumpla la condición

while e==False:
  try:
    #aqui va ocurrir el error....................
    tamaño=int(input("Ingrese el tamaño de la lista: "))
    #..................................................
    for i in range(tamaño):
      print("num",i)
      nm=input("ingrese el nombre del empleado/a ")

      empleados.append(nm)
    print(empleados)
    e=True

  except ValueError:
    print("ERROR: por favor, digita únicamente valores numéricos enteros")
```

```python
print()
print("MENÚ")
print("1:sumar")
print("2:restar")
print("3:multiplicar")
print("4:dividir (/)")
print()
#ingresando datos:
opción= input("Ingrese la opción deseada del menú: ")
numero_uno=float(input("Ingrese el primer numero de la operación: "))
numero_dos=float(input("Ingrese el segundo numero de la operación: "))
if opción=="1" or opción=="suma" or opción=="+":
  print("se suman los valores ingresados")
  print("la suma es:",numero_uno+numero_dos)
if opción=="2" or opción=="resta" or opción=="-":
  print("se resta los dos valores ingresados")
  print("la resta es:",numero_uno-numero_dos)
if opción=="3" or opción=="multiplicación" or opción=="*":
  print("vamos a multiplicar los valores ingresados")
  print("la multiplicación es:",numero_uno*numero_dos)
if opción=="4" or opción=="división" or opción=="/":
  print("vamos a dividir los valores ingresados")
  if numero_dos!=0:
    print("la división es:",numero_uno/numero_dos)
  else:
    print("¡no es posible dividir entre cero!")
```

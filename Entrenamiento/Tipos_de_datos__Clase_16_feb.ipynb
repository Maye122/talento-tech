#2) LISTAS

Usa llaves [ ],
se pueden modificar

```python
Nombres=["Daniel","Maye","sofia"]
edades=[24,30,40]
varios=["cc",233,"Talla"]
print(Nombres)
print(edades)
print(varios)
```

```python
#Para cambiar el valor de una lista
Nombres[0]="Lau"
print(Nombres)

#Agregar otro valor en la lista en la posición deseada
edades.insert(3,20)
print(edades)

#Agregar valor en la ultima posicion de la lista
Nombres.append("Eider")
print(Nombres)

#para remover
Nombres.remove("sofia") #si hay valores repetidos elimina el de la primera posición
print(Nombres)

edades.remove(20)
print(edades)

#Eliminar valores en la posición que uno quiera
#eliminar edades por posición fn pop y remove
edades.pop(0)
edades.remove(edades[0])
print(edades)



```

#4) INPUT
Se usa para pedir datos al usuario

```python
nombre=input("Ingresa tu nombre: ")
print("Hola,",nombre)

#convierte el número tipo texto a tipo flotante
edad=float(input("Ingresa tu edad:"))
print("tu edad es", edad)

#para redondear
op=round(edad/7,3)
print("op: ", op)

#otra forma es dandole formato a la variable
op=edad/7
print(f"{op:.2f}")
```

```python
notas=[3.5,4,5,2,1,1.6,4.7]
prom=sum(notas)/len(notas)  #len es para contar la cantidad de elementos
print(f"el promedio de notas es: {prom:.2f}")
```

```python
#import libraries
import pandas as pd  #sirve para trabajar con datos (tablas tipo excel)
import matplotlib.pyplot as plt  #sirve para hacer gráficas
import numpy as np  #para cálculos matemáticos y arreglos

#Por ahora solo se va usar pandas, no importa si se importan varias y no se usan


#Crear la base datos
#Construir las listas de valores

value_n_list=["andrew","philip","Jhon"]
print(value_n_list)

value_w_list= [50, 78, 91.5] #kg
value_h_list=[1.76, 1.87, 1.69] #m


#Usar un diccionario de la forma ({key:value})
#con el formato ("str":list)

data_dict={"name": value_n_list,
           "weight": value_w_list,
           "height": value_h_list}
print(data_dict)

#convierte el diccionario en un marco de datos (tabla de datos)
df= pd.DataFrame(data_dict)
print(df)
df


#Luego para que salga la tabla presentada bonita hay que llamar la tabla
print()
df

```
